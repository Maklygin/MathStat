### Известно, что электричка "Вашингтон-Петушки" аварийно останавливается раз в несколько дней. Аналитики РЖД проанализировали, сколько дней электричка едет без поломок, и составили выборку: x = (3, 22, 13, 6, 18, 5, 6, 10, 7, 15).

РЖД хочет проверить гипотезу, что дисперсия распределения равна 9 против правосторонней альтернативы.

Требуется:

Ввести предположение, каким распределением описывается данная выборка.

Проверить гипотезу о значении дисперсии распределения для уровня значимости alpha = 0.05 аналитически.
Вывести и получить доверительный интервал для значения дисперсии при alpha = 0.05.
Все выкладки должны быть сделаны аналитически, без использования компьютера. (допускается использование компютера для подстановвки численых значений в финальную формулу)



```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
x = np.array([3, 22, 13, 6, 18, 5, 6, 10, 7, 15])
n = len(x)

# оценим, может ли выборка иметь нормальное распределение 
qq_points, qq_line =st.probplot(x, plot=plt)
plt.show()

plt.hist(x)
plt.show()
```

1.   Предположим, что выборка распределена нормально.
2.   $X$ $\in$ N($\mu,\sigma^2$) 

     $H_{0}: \sigma_0 = 9$

     $H_{1}: \sigma > \sigma_0 $
3.   Критерий хи-квадрат:
     
     Статистика $T = \frac{(n-1)S^2}{\sigma_0^2} \in χ^2(n-1)$
4.   Проверка гипотезы (alpha = 0.05):
    
     $$ S^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar X)^2  $$

     $$ \mathbb{P} (T \geq t_{1-α}| T \sim \chi_{n-1}) = \alpha $$
     Табличное значение квантиля хи-квадрат $\chi^9_{0.95} = 16.919$, тогда как значение статистики T = 4.37.
     Следовательно гипотезу не отклоняем.



    


```python
sigma = 9.
X_mean = np.sum(x)/n
S_square = (1/(n-1))*( np.sum((x-X_mean)**2) ) 
T = (n-1)*S_square/sigma**2
print(T)
```

5.    Построим доверительный интервал для дисперсии.

Так как случайная величина $ \frac{(n-1)S^2}{\sigma^2}$ имеет распределение $\chi(n-1)$, то выразив отсюда дисперсию получим:

Табличные данные:

$\chi^9_{0.975} = 19,0228$ 

$\chi^9_{0.025} = 2,7004$	

```python
print(((n-1)*S_square/19.0228)**0.5, ((n-1)*S_square/2.7004)**0.5)
```

Как мы видим дисперсия гипотезы попадает в 95% доверительный интервал.
